```python
import pandas as pd

#Data Cleaning

df = pd.read_csv("T1.csv")

print(df.head())
print(df.info())
print(df.describe())

df.isnull().sum() #Checking for nulls
```

No nulls, data is already quite clean

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.scatter(
    df['Wind Speed (m/s)'],
    df['LV ActivePower (kW)'],
    alpha=0.2,
    s=5
)

plt.xlabel("Wind Speed (m/s)")
plt.ylabel("Actual Power (kW)")
plt.title("Wind Turbine Power Curve (Actual)")

plt.show()
```
